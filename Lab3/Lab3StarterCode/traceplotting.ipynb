```python
import matplotlib.pyplot as plt
import json

file = "data/algo2.json"
with open(file, "r") as f:
    performance = json.load(f)
print(performance.keys())

# Sample data
categories = []
quality = []
rebuffer_time = []
variation = []
qoe = []
for test in performance.keys():
    if test != "overall":
        categories.append(test)
        qoe.append(performance[test]["QoE"])
        quality.append(performance[test]["Total Quality"])
        rebuffer_time.append(performance[test]["Rebuffer Time"])
        variation.append(performance[test]["Total Variation"])
```

```python
# Creating bar graph
plt.bar(categories, qoe)
plt.xticks(rotation=45, ha='right')

# Adding title and labels
plt.title('QOE of Traces')
plt.xlabel('Trace (Throughput_Variance)')
plt.ylabel('Average QOE')

# Displaying the graph
plt.show()
```

```python
plt.bar(categories, quality)
plt.xticks(rotation=30, ha='right')

# Adding title and labels
plt.title('Total Quality of Traces')
plt.xlabel('Trace (Throughput_Variance)')
plt.ylabel('Total Quality')

# Displaying the graph
plt.show()
```

```python
plt.bar(categories, variation)
plt.xticks(rotation=30, ha='right')

# Adding title and labels
plt.title('Total Variation of Traces')
plt.xlabel('Trace (Throughput_Variance)')
plt.ylabel('Total Variation')

# Displaying the graph
plt.show()
```

```python
plt.bar(categories, rebuffer_time)
plt.xticks(rotation=30, ha='right')

# Adding title and labels
plt.title('Rebuffer Time of Traces')
plt.xlabel('Trace (Throughput_Variance)')
plt.ylabel('Rebuffer Time')

# Displaying the graph
plt.show()
```
