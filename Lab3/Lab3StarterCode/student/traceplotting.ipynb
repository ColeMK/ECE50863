```python
import matplotlib.pyplot as plt

# Sample data
categories = ['lo_lo', 'lo_mi', 'lo_hi', 'mi_lo','mi_mi', 'mi_hi', 'hi_lo', 'hi_mi', 'hi_hi']
quality = [72, 48, 124, 185, 130, 197, 217, 217, 214]
rebuffer_time = [12.58, 1, 2.12, .5, 7.22, .4, .47, .33, .33] 
variation = [17, 24, 19, 13, 12, 19, 15, 3, 3]
qoe = [.11, .268, .887, 1.477, .796, 1.556, 1.738, 1.792, 1.767]
# Creating bar graph
plt.bar(categories, qoe)

# Adding title and labels
plt.title('QOE of Traces')
plt.xlabel('Trace (Throughput_Variance)')
plt.ylabel('Average QOE')

# Displaying the graph
plt.show()
```
